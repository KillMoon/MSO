## 1 пара
Выдвинем гипотезу, что мат ожидания 1 пары равны.

```python
sns.displot((data.couple_1_1, data.couple_1_2), kde=True, bins = 20)
```

Парралельно выдвинем гипотезу что дисперсии равны

Проверим выборки на нормальность с помошью критерия Шапиро-Уилка.
Найдём их мат ожидания, проанализируем

```python
p = stats.shapiro(data.couple_1_1)
print(p[1])
p = stats.shapiro(data.couple_1_2)
print(p[1])

m1 = data.couple_1_1.mean()
m2 = data.couple_1_2.mean()
print(m1, m2)

l1 = len(data.couple_1_1)
l2 = len(data.couple_1_1)

print(l1, l2)

a = 0.05
```

```python
p1 = stats.fligner(data.couple_5_1, data.couple_5_2)
print("p = " + str(p1[1]))
if p1[1] > a:
    print("Гипотеза принята")
else:
    print("Гипотеза отвергнута")
```

Как видно мат ожидания сильно различаются и выборки не распределены по норм закону, выборка не большая => применяем критерий Манна-Уитни

```python
p1 = stats.mannwhitneyu(data.couple_1_1, data.couple_1_2)
print("p = " + str(p1[1]))
if p1[1] > a:
    print("Гипотеза принята")
else:
    print("Гипотеза отвергнута")
```

```python
p_array = []

for p in np.arange(0, 1, 0.01):
  arr = np.random.choice(2, 100, p=[p, 1-p])
  tmp_arr = []
  for x in arr:
    if x == 0:
      tmp_arr.append(np.random.normal())
    elif x == 1:
      tmp_arr.append(np.random.weibull(1 - p))
  p_array.append(stats.shapiro(tmp_arr)[1])

p_array = np.array(p_array)
dataframe = pd.DataFrame(p_array, columns=['p_arr'])
dataframe.plot.density()
dataframe.plot.hist()
sns.displot(p_array, kde=True, bins=20)
```
